```python
## 반복문으로 팩토리얼 구하기
def factorial(n):
  result =1
  for i in range(1,n+1):
    result*=i

  return result

n = 5
print('%d!=%d'%(n, factorial(n)))

#================================================================
#  재귀함수: 자기 자신을 호출하는 것
#================================================================
# (예제) 팩토리얼 구하기 factorial(0)=1, factorial(n)=n*factorial(n-1) :
def factorial2(n):
  if n==0:
    return 1
  else:
    return n*factorial2(n-1)

n = 5
print('%d!=%d'%(n, factorial2(n)))

#================================================================
## (예제) 독일 수학자 콜라츠(Collatz -> L.) 추측: 주어진 숫자가 짝수면 2로 나눈다.
# 주어진 숫자가 홀수면 3배한 후 1을 더한다.
def collatz(num):
  if num==1:
    return 1
  elif num%2==0:
    print(num, end='->')
    return collatz(num//2)
  else:
    print(num, end='->')
    return collatz(3*num+1)

collatz(7)
collatz(128)
collatz(129)

#시행횟수============================================================
def collatz_count(num):
  if num==1:
    return 0
  elif num%2==0:
    return collatz_count(num//2)+1
  else:
    return collatz_count(3*num+1)+1

print(collatz_count(7), "회")
print(collatz_count(128), "회")
print(collatz_count(129), "회")

#================================================================
# (예제) 피보나치 수열: 1, 1, a_n+2=(a_n+1)+(a_n)
def fibonacci(n):
  if n==1 or n==2:
    return 1
  else:
    return fibonacci(n-1)+fibonacci(n-2)

n=10
print('fibonaci(%d)=%d'%(n, fibonacci(n)))

# n=50일때 1시간 넘게 걸림!!

counter = 0
def fibonacci2(n):
  global counter#전역변수
  counter +=1

  if n==1 or n==2:
    return 1
  else:
    return fibonacci2(n-1)+fibonacci2(n-2)


n=10
fibonacci2(n)

print('fibonacci({}) 계산에 활용된 덧셈 횟수는 {}번입니다.'.format(n,counter))

# fibonacci(10) 계산에 활용된 덧셈 횟수는 109번입니다.
# => 같은 값을 구하는 연산을반복하고 있기 때문에 발생!!!

#================================================================
## 메모화: 같은 값을 한번만 계산하도록 코드를 수정하기!
#================================================================

dict={1:1,2:1}
def fibonacci_memo(n):         #global dict 있어도되고 없어도됨
  if n in dict:
    return dict[n]
  else:
    output = fibonacci_memo(n-1)+fibonacci_memo(n-2)
    dict[n] = output
    return output

n=50
print('fibonaci(%d)=%d'%(n, fibonacci_memo(n))) # 엄청빠름!!

# Exercise (facotiral 메모화) ==========================================
dict={0:1}
def factorial_memo(n):
  if n in dict:
    return dict[n]
  else:
    output = n*factorial_memo(n-1)
    dict[n] = output
    return output


n=10
print(factorial_memo(n))
```
