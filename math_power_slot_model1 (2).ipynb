```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
  if 1<= x <=20:
      return 50 * 110
  if 20< x <=40:
      return 100 * 110
  if 40< x <=63:
      return 200 * 110
  if 63< x <=75:
      return 300 * 110
  if 75< x <=108:
      return 200 * 110
  if 108< x <=130:
      return 100 * 110
  if 130< x:
      return 10 * 110

def g(x):
    if 1<= x <=20:
        return 50*x
    if 20< x <=40:
        return 50*20 + 100*(x-20)
    if 40< x <=63:
        return 50*20 + 100*20 + 200*(x-40)
    if 63< x <=75:
        return 50*20 + 100*20 + 200*23 + 300*(x-63)
    if 75< x <=108:
        return 50*20 + 100*20 + 200*23 + 300*12 + 200*(x-75)
    if 108< x <=130:
        return 50*20 + 100*20 + 200*23 + 300*12 + 200*33 + 100*(x-108)
    if 130< x:
        return 50*20 + 100*20 + 200*23 + 300*12 + 200*33 + 100*22 + 10*(x-130)
x = np.linspace(0, 132, 1001)

fx = np.zeros_like(x)
gx = np.zeros_like(x)
difference = np.zeros_like(x)

for i in range(1001):
    fx[i] = f(x[i])
    gx[i] = g(x[i])
    difference[i] = fx[i] - gx[i]

# Check for large jumps in difference to determine discontinuity
jumps = np.where(np.abs(np.diff(difference)) > 1000)[0]
difference[jumps] = np.nan
difference[jumps + 1] = np.nan

# Plotting f(x) and g(x) on the same plot
plt.figure(figsize=(16, 9))
plt.plot(x, fx, 'r.', label='f(x)')
plt.plot(x, gx, 'b.', label='g(x)')
plt.title('Plot of f(x) and g(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()

# Plotting f(x) - g(x)
plt.figure(figsize=(16, 9))
plt.plot(x, difference, 'g-', label='f(x) - g(x)')
plt.title('Plot of f(x) - g(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```
