```python
## 예제 5: 함수의 미분 ===============================
# forward difference: D_{+}f = (f(a+h)-f(a))/h
# backward difference: D_{-}f = (f(a)-f(a+h))/h
# central difference: D_{0}f = (f(a+h)-f(a+h))/(2*h)

#f=x**3+4*x**2+5*x
#df = 3x^2+8X+5
import numpy as np
import math
import matplotlib.pyplot as plt
f = lambda x : x**3+4*x**2+5*x #익명함수이용
df = lambda x : 3*x**2+8*x+5

x_0=-3; x_n=3; n=100
h= (x_n-x_0)/n
x=np.linspace(x_0, x_n, n)#n+1로 나누어야한다.중요(n+1이라고 쓰면 n칸으로 나누어준다.)
Dpf = list(map(lambda a : (f(a+h)-f(a))/h, x))#단일값이 입력되는 것이 아니기떄문에 map함수해주었다.
Dmf = list(map(lambda a : (f(a)-f(a-h))/h, x))
D0f = list(map(lambda a : (f(a+h)-f(a-h))/(2*h), x))
df_exact = list(map(df, x))

plt.plot(x, Dpf, label = '$D_{+}f$')
plt.plot(x, Dmf, label = '$D_{-}f$')
plt.plot(x, D0f, label = '$D_{0}f$')
plt.plot(x, df_exact, label = '$df$')
plt.legend()
plt.show()
```

```python
#Euler method로 미분방정식의 해를 찾는것을 코딩한 것이다.
# y'=f(t,y)
# y'=y, y(0)=1
t0 = 0; tn = 1; n = 10
t = np.linspace(t0, tn, n+1) #1차원 벡터배열 n개
dt = (tn-t0)/n
y = np.zeros_like(t) #t 1차원 벡터배열

f = lambda t, y: y #y'=y이기 때문에

y[0]=1#초깃값(initial value)
for i in range(len(t)-1) : #n-1번 해야함을 알아야한다.
  y[i+1] = y[i] + dt*f(t[i],y[i])#y의 미소증가량만큼 증가해 나간다. 배열의 i번째를 증가한것으로 병경하는것이다.


plt.plot(t,y, label='Euler method')
plt.plot(t, np.exp(t), label='exact')
plt.legend()
plt.show()
```

```python
## Differential equation:
# y'=f(t,y), y (t0) =y0:
# Euelr's method: y_{n+1}=y_n + dt*f(tn, yn)
# e.g. y'=y, y(0)=1: => t=1일 때, y값은?
import numpy as np
import matplotlib.pyplot as plt

t0 = 0
tn = 1
n = 100
t = np. linspace (t0, tn, n+1) # (n+1)개의 array, t0, t1, t2... tn
dt= (tn-t0)/n #t의 간격=0.01
y = np.zeros_like (t)#0으로 배열만들고 n번째마다 변경하는 형식으로 진행된다.

t0 = 0
tn = 1
n = 100
dt= (tn-t0)/n #t의 간격=0.01
t = np.arange (t0, tn+dt, dt) # t0, t1 , ..., tn(arrange는 숫자배열을 만드는 함수이다.)
y = np.zeros_like(t)

## e.g. y'=y, y(0)=1: => exact solution: y=e^t
f = lambda t, y: y#이 식이 y'=y이기떄문에 y만 쓰면된다.
y_exact = np.exp(t)

y[0]=1 #문제에서 주어진 초기조건
for i in range (len(t)-1):
  y[i+1] = y[i] + dt*f(t[i], y[i])

plt.plot(t,y,'r.',label="Euler's method")#빨간색 점으로 그리는 포맷이다.
plt.plot(t, y_exact, label='exact solution')
plt.legend()
plt.show()
#=================================================================

# y'=cos(t), y(0)=0 => y_exact=sin(t)
#=> t=2*pi 일떄 y=?
import numpy as np
import matplotlib.pyplot as plt
t0 = 0
tn = 2*np.pi
dt = 0.1
t = np.arange(t0, tn+dt, dt) #numpy모듈 사용할때는 이렇게 array배열로 만들어주어서 함수만든다.
y = np.zeros_like(t)

f=lambda t, y: np.cos(t)
y_exact = np.sin(t)

y[0]=0 #초기조건
for i in range(len(t)-1):
  y[i+1] = y[i] + dt*f(t[i], y[i])#미분방정식에서 반복적을 나오는 원리코드이다.

plt.plot(t,y,'r.', label="Euler's method")
plt.plot(t,y_exact, label='exact solution')
plt.legend()
plt.show()
#==================================================================

#y'+2y=2-e^{-4t}, y(0)=1
#Use Euler's Method with a step size of dt = 0.1, 0.05, 0.01
#to find approximate values of the solution at t=1일떄 y=? (t=0-t=1일떄 y값은?).
#Compare them to the exact values of the solution at these points.

t0 = 0
tn = 1
f = lambda t,y: -2*y+2-np.exp(-4*t) #미분방정식을 쓰는 곳이다.변수가 t,y두개라는 뜻이다.(미분꼴을 쓴다)
f_exact = lambda t: 1+0.5*np.exp(-4*t)-0.5*np.exp(-2*t)

#dt=0.1일때 plot
dt = 0.1
t = np.arange(t0, tn+dt, dt)
y = np.zeros_like(t)
y[0]=1
for i in range(len(t)-1):
  y[i+1]=y[i]+dt*f(t[i],y[i])

plt.plot(t,y,'r.',label="Euler dt="+str(dt))
plt.plot(t, f_exact(t),label='exact solution')
plt.legend()
plt.show()

#dt=0.01일때 plot
dt = 0.01
t = np.arange(t0, tn+dt, dt) #t0,...,tn
y = np.zeros_like(t)
y[0]=1 #초기조건
for i in range(len(t)-1):
  y[i+1] = y[i] + dt*f(t[i], y[i])
plt.plot(t,y,'r.',label="Euler dt="+str(dt))
plt.plot(t, f_exact(t),label='exact solution')
plt.legend()
plt.show()
```

```python
## 예제 6: 함수의 적분 ===============================
#상적분 (upper integral)
#하적분 (lower integral)
# x=0-x=pi 까지 f=sin(x) 함수 적분
import numpy as np
x0 = 0
xn = np.pi
n = 100
x = np.linspace(x0,xn,n+1) #주의하세요!! (n+1)-dimensional array
dx = (xn-x0)/n
exact = 2 #exact integral value
f = np.sin(x) # (n+1)-dimensional array

lower_sum = sum(f[0:n])*dx
upper_sum = sum(f[1:])*dx
error_lower = lower_sum-exact
error_upper = upper_sum-exact
print(error_lower)
print(error_upper)


#Trapezoidal(사다리꼴) rule: 사다리꼴로 나누어 정적분
len(f)
trap_sum = 0.5*dx*(f[0]+2*sum(f[1:n]+f[n]))
error_trap = trap_sum - exact
print(error_trap)

# Simpson's rule: [a,b] 구간을 n등분한 뒤, 각 구간을 2차 함수로 근사한 후, 2차 함수의 넓이를 더해 적분함
# int_a^b f(x) dx = 1/3*h*[Sigma_{i=1}^{n/2} f(x_{2i-2})+4*f(x_{2i-1})+f(x_{2i})]
#                 = 1/3*h*[f(0)+2*sigma_{i=1}^{n/2-1}f(x_{2i}) +4*sigma_{i=1}^{n/2}f(x_{2i-1})+f(x_n)]

tmp = f[0]+2*sum(f[2:n-1:2])+4*sum(f[1:n:2])+f[n]
sim_sum = (dx/3)*tmp
error_sim = sim_sum-exact
print(error_sim)
```
