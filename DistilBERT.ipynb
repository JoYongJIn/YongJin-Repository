```python
import transformers

print(transformers.__version__) # transformers의 버전을 확인합니다.
```

```python
from transformers import pipeline

senti = pipeline('sentiment-analysis') # Hugging Face의 transformers라이브러리를 사용하여 감성분석 파이프라인을 생성합니다.
```

```python
senti.model # 'DistilBertForSequenceClassification' 모델의 구조를 보여줍니다. 텍스트 시퀀스를 분류하기 위한 모델입니다.
```

```python
# 문자예시를 통해 감성분석을 해봅니다.
print(senti(["The book was okay, but it didn't quite meet my expectations. Some chapters felt unnecessary."]))
```

```python
print(senti(["I can't decide whether I love or hate this song; it's both catchy and annoying at the same time."]))
```
